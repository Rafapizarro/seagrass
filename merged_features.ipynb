```python
import pandas as pd
```

```python
GCP_PROJECT = "seagrass-lewagon"
GCP_REGION="europe-west1"
BQ_DATASET = "seagrass"
```

```python
query = f"""
    SELECT *
    FROM `{GCP_PROJECT}.{BQ_DATASET}.merged_features`
    """
```

```python
! pwd
```

```python
from google.cloud import bigquery

client = bigquery.Client(project=GCP_PROJECT)
query_job = client.query(query)
result = query_job.result()
merged_features = result.to_dataframe()
merged_features.to_csv("data/merged_features.csv", header=True, index=False)
```

```python
merged_features = pd.read_csv("data/merged_features.csv")
merged_features
```

```python
import folium
```

```python
# Define the center of the Mediterranean for the map
mediterranean_center = [34.5, 18.0]  # Approximate central coordinates

# Create a Folium map
m = folium.Map(location=mediterranean_center, zoom_start=5)

# Add points to the map
for _, row in merged_features.iterrows():
    folium.CircleMarker(
        location=[row["lat"], row["lon"]],
        radius=2,  # Small dot for each point
        color="blue",
        fill=True,
        fill_color="blue",
        fill_opacity=0.6
    ).add_to(m)

# Save and display the map
m.save("mediterranean_map.html")
print("✅ Map saved as 'mediterranean_map.html'. Open it in a browser to view.")
```
